Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def experience_to_months(exp):
    exp = str(exp)

    if exp == 'nan' or 'Не указано'  in exp:
        return np.nan

    years = 0
    months = 0

    if 'лет' in exp or 'год' in exp:
        year_part = exp.split('лет')[0].split('год')[0].split()[-1]
        if year_part.isdigit():
          if int(year_part) < 2000:
            years = int(year_part)

    if 'месяц' in exp:
        month_part = exp.split('месяц')[0].split()[-1]
        if month_part.isdigit():
            months = int(month_part)

    return years * 12 + months


data['Опыт работы (месяц)'] = data['Опыт работы'].apply(experience_to_months)

data = data.drop('Опыт работы', axis=1)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань',
                  'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа',
                  'Красноярск', 'Пермь', 'Воронеж','Волгоград']

def city_moving(text):
    text = str(text)
    parts = [part.strip() for part in text.split(',')]

    # Город
    city = parts[0].split('м.')[0].strip() if 'м.' in parts[0] else parts[0].strip()

    if city == 'Москва':
        city_category = 'Москва'
    elif city == 'Санкт-Петербург':
        city_category = 'Санкт-Петербург'
    elif city in million_cities:
        city_category = 'город-миллионник'
    else:
        city_category = 'другие'


    if 'м.' in parts[1]:
      parts = parts[2::]
    else:
      parts = parts[1::]



    moving = False
    # Готовность к переезду
    part_lower = parts[0].lower()
    if 'не' not in part_lower and part_lower:
      moving = True


    # Готовность к командировкам
    business_trips = False

    part_lower = parts[-1].lower()
    if 'не' not in part_lower and part_lower:
        business_trips = True



    return city_category, moving, business_trips


data[['Город', 'Готовность к переезду', 'Готовность к командировкам']] = data['Город, переезд, командировки'].apply(
    lambda x: pd.Series(city_moving(x))
)


data = data.drop('Город, переезд, командировки', axis=1)
```

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Итак, давайте обсудим возможный алгоритм преобразования:
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
currency_map = {
    'руб': 'RUB', 'рублей': 'RUB', 'рубля': 'RUB', 'р.': 'RUB', 'р': 'RUB',
    'USD': 'USD', 'долларов': 'USD', 'доллара': 'USD', '$': 'USD',
    'EUR': 'EUR', 'евро': 'EUR', '€': 'EUR',
    'грн': 'UAH', 'гривен': 'UAH', 'гривны': 'UAH',
    'бел.руб.': 'BYN', 'белруб': 'BYN',
    'KZT': 'KZT', 'тенге': 'KZT',
    'сум': 'UZS'
}


proportion_map = {
    'RUB': 1, 'USD': 1, 'EUR': 1, 'UAH': 10, 'BYN': 1, 'KZT': 100, 'UZS': 10000
}


exchange_rates = pd.read_csv('/content/ExchangeRates.csv')
exchange_rates['date'] = pd.to_datetime(exchange_rates['date'], format='%d/%m/%y').dt.date

# Функция для разделения суммы и валюты
def extract_salary(salary_str):
    if pd.isna(salary_str):
        return None, 'RUB'

    parts = str(salary_str).split()
    amount = float(''.join(filter(str.isdigit, parts[0]))) if parts else None


    currency = 'RUB'  # по умолчанию
    for part in parts:
        if part.lower() in currency_map:
            currency = currency_map[part.lower()]
            break

    return amount, currency


data[['salary_amount', 'salary_currency']] = data['ЗП'].apply(
    lambda x: pd.Series(extract_salary(x))
)


data['proportion'] = data['salary_currency'].map(proportion_map)
data['date'] = pd.to_datetime(data['Обновление резюме'], format='%d.%m.%Y %H:%M').dt.date


data = data.merge(
    exchange_rates[['currency', 'date', 'close']],
    left_on=['salary_currency', 'date'],
    right_on=['currency', 'date'],
    how='left'
)

# Заполняем пропуски для рубля
data['close'] = data['close'].fillna(1.0)
data['proportion'] = data['proportion'].fillna(1.0)


data['ЗП (руб)'] = data['salary_amount'] * data['close'] / data['proportion']
data = data.drop(['salary_amount', 'salary_currency', 'proportion', 'currency', 'date'], axis=1)
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
data['log_возраст'] = np.log(data['Возраст'])

plt.figure(figsize=(12, 6))
histplot = sns.histplot(data['log_возраст'], kde=True, bins=30)
plt.title('Распределение возраста в логарифмическом масштабе')
plt.xlabel('Логарифм возраста')


log_mean = data['log_возраст'].mean()
log_std = data['log_возраст'].std()

histplot.axvline(log_mean, color='k', lw=2, label=f'Среднее: {np.exp(log_mean):.1f} лет')
histplot.axvline(log_mean - 3*log_std, color='r', lw=2, linestyle='--', label='-3σ')
histplot.axvline(log_mean + 4*log_std, color='r', lw=2, linestyle='--', label='+4σ')

plt.legend()
plt.show()


z_scores = (data['log_возраст'] - log_mean) / log_std
outliers = data[z_scores > 4]

print(f"Количество выбросов по возрасту (z > 4): {len(outliers)}")
print(f"\nГраница выбросов: > {np.exp(log_mean + 4*log_std):.1f} лет")

if len(outliers) > 0:
    print(f"\nТаблица выбросов:")
    print(outliers[['Возраст', 'log_возраст']].sort_values('Возраст', ascending=False))


initial_count = len(data)
data_cleaned = data[z_scores <= 4]
final_count = len(data_cleaned)

print(f"\nУдалено выбросов по возрасту: {initial_count - final_count}")
print(f"Осталось записей: {final_count}")

data = data_cleaned
```

Форма распределения:

    Распределение асимметрично в левую сторону (левый хвост длиннее)

    длинный левый хвост - молодые соискатели

Положение линий:

    Черная линия (среднее) - показывает средний логарифм возраста

    Красные пунктирные линии - границы ±3σ/+4σ для выявления выбросов

    Правая граница +4σ определяет пожилых соискателей-выбросов

Интерпретация:
График показывает, что после логарифмирования распределение стало более нормальным, но сохранило асимметрию, характерную для возрастных данных на рынке труда.
